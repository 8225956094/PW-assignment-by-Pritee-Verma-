6. Python code for the given number of days, find how many weeks and how many days are left
### eg. input = 15 days, # weeks =  2, # days left = 1

```python
a=int(input("How many days are left:"))
b=int(input("How many weeks are left:"))
days_left=(a)-(b*7)
print("Days left:",days_left)
```

### 7. Python code for the given number of days, find how many weeks and how many days are left
### eg. input = 15 days, # weeks =  2, # days left = 1

```python
a=int(input("How many days are left:"))
b=a//7
c=a%7
print("How many weeks are left:",b)
print("How many days are left:",c)
```
