**5.What is mutability in Python?**

**Definition:**

Mutability means whether the value of an object can be changed after it is created.

If an object can be changed, it is called mutable.

**Examples:**

 Mutable objects → You can change their content.

list, set, dict

**6. Why are lists mutable, but tuples are immutable?**

- Lists are mutable because they are designed to store and manage dynamic data.

You can add, remove, or change elements after the list is created.

- Tuples are immutable because they are meant to store fixed data that should not change.

Once created, the elements inside a tuple cannot be modified.

**12. How does the elif statement work?**

**Definition:**

The elif (else if) statement is used to check multiple conditions one by one after an if statement.

**How it works:**

1. Python first checks the if condition.


2. If if is False, then it checks the next elif condition.


3. If one elif is True, that block runs and the rest are skipped.


4. If none of the conditions are True, the else block runs.

```python

#Example
marks = int(input("Enter your marks: "))

if marks >= 90:
    print("Grade: A")
elif marks >= 75:
    print("Grade: B")
elif marks >= 50:
    print("Grade: C")
else:
    print("Grade: Fail")
```

**13. What is the difference between for and while loops?**

**Difference between for and while loops in Python**

1. For Loop

Used when we know the number of repetitions in advance.

Works with sequences like lists, strings, or ranges.

Example use: printing numbers from 1 to 10.



2. While Loop

Used when we don’t know exactly how many times the loop should run.

Keeps running until the condition becomes False.

Example use: keep asking for a password until it’s correct.

**14. Describe a scenario where a while loop is more suitable than a for loop.**

**Scenario where a while loop is better:**

A while loop is more suitable when we don’t know in advance how many times the loop should run, and it depends on a condition.

**Example scenario:**

Asking a user to enter the correct password.

The loop keeps running until the user types the right password.

Here, we cannot fix the number of attempts, so a while loop is best.

**13. Write a Python program to find the sum of all even numbers between 1 and 50.**

```python
#the sum of all even numbers between 1 and 50.

sum_even = 0
for i in range (1, 51):
    if (i % 2 == 0):
        sum_even += i

print("The sum of all even numbers between 1 and 50 is:", sum_even)
```

**14. Write a program to reverse a string using a while loop.**

```python
#A program to reverse a string using a while loop.

String = "PRITI"

Reverse_String = ""

#find length of String
Length = len(String) - 1

while Length >= 0:
    Reverse_String += String[Length]
    Length -= 1

print("Reverse_String: ",Reverse_String)
```

**15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.**

```python

#a Python program to calculate the factorial of a number provided by the user using a while loop.

Number = int(input("Please enter a number: "))

factorial = 1
i = 1

while i <= Number:
    factorial = factorial * i
    i = i + 1

print("Factorial of", Number, "is", factorial)
```
